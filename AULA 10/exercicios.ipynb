```python
# Exercicio 1

# Crie 3 listas:
# ▶ Inteiros: a primeira lista com 10 números inteiros gerados
# aleatoriamente
# ▶ Reais: a segunda lista com 5 números reais gerados
# aleatoriamente
# ▶ Strings: A terceira lista com 7 strings criadas por você.
# Então adicione as 3 listas a uma lista única, chamada
# completa.
# Apague todas as 3 listas originais.
# Acesse e mostre todos os elementos da lista completa.

# Use:
# from random import randint, random
# print(randint(0, 10))
# print(random() * 10)


from random import randint, random

print(randint(0, 10))
print(random() * 10)

numero_inteiro = []
numero_real = []

for i in range (10):
    n = randint(1,100)
    numero_inteiro.append(n)
print(numero_inteiro)

for i in range (5):
    n = random() * 100
    numero_real.append(n)
print(numero_real)

string = ["a", "b", "c", "d", "e", "f", "g"]
print(string)

listas = [numero_real[:], numero_inteiro[:], string[:]]
print(listas)

del numero_inteiro
del numero_real
del string


print(listas)
```

```python
def ler_matriz():
    matriz = []
    for linha in range(3):  # 3 linhas
        nova_linha = []
        for coluna in range(3):  # 3 colunas
            numero = int(input("Digite um numero: "))
            nova_linha.append(numero)
        matriz.append(nova_linha)
    return matriz

def mostrar_matriz(matriz):
    print("\nMatriz:")
    for linha in matriz:
        for valor in linha:
            print(f"{valor:2}", end=" ")
        print()

def soma_diagonal_principal(matriz):
    soma = 0
    for linha in range(3):
        soma += matriz[linha][linha]  # Diagonal principal: linha == coluna
    return soma

def main():
    matriz = ler_matriz()
    mostrar_matriz(matriz)
    soma = soma_diagonal_principal(matriz)
    print(f"\nSoma da diagonal: {soma}")

main()
```
