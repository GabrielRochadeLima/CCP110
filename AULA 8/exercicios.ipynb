```python
# Exercício 3

lista = []
for i in range (10):
    n = int(input("Digite um numero: "))
    lista.append(n)
    
    
maior = lista [0]
indice_do_maior = 0
for indice in range(len(lista)):
    # print(lista[indice], indice)
    if lista[indice] > maior:
        maior = lista[indice]
        indice_do_maior = indice
       
print(f"A lista completa é: {lista}")
print(f"O maior valor é {maior} no indice {indice_do_maior}")
```

```python
# Exercicio 4

# Faça um programa para criar uma lista de 10 elementos do usuário e aprsente:
#  a soma dos elementos pares
#  a soma dos elementos de índice par

lista = []
for i in range (10):
    n = int(input("Digite um numero: "))
    lista.append(n)

soma_pares = 0
soma_incice_pares = 0

for indice in range(len(lista)):  
    # elemento que esta no indice 
    if lista[indice] % 2 == 0:
        soma_pares += lista[indice]
    
    # soma elemento que esta no indice par
    if indice % 2 == 0:
        soma_incice_pares += lista[indice]

print(f"A lista completa é: {lista}")
print(f"A soma dos pares é: {soma_pares}")
print(f"A soma indice par é: {soma_incice_pares}")
    
```

```python
# Exercicio 5

lista = []

for i in range (5):
    n = int(input("Digite um número"))
    lista.append(n)

print(lista)
lista_reversa = lista[::-1]
print("Inverso: ")
print(lista_reversa)

for elemento in lista_reversa:
    print(elemento)
```

```python
# Exercicio 7

temperaturas = [-10, -8, 0, 1, 2, 5, -2, -4]

menor = min(temperaturas)
maior = max(temperaturas)
media = sum(temperaturas) / len(temperaturas)

print(f"A menor temperatura é: {menor}")
print(f"A maior temperatura é: {maior}")
print(f"A media da temperatura é: {media}")

        
```
