```python
# Exercicio 1

dicionario = {
    'alpha': 1, 
    'bravo': 2, 
    'charlie' : 1, 
    'delta': 3, 
    'echo': 1
}

def procurarChave (dicionario, valorProcurado):
    lista= []
    for chave, valor in dicionario.items():
        if valor == valorProcurado:
            lista.append(chave)
    return lista

print(procurarChave(dicionario, 1))
```

```python
# Exercicio 2

from random import randint

lista = []

for i in range(100):
    lista.append(randint(0,20))
    
print(lista)

dicionario = {}

for numero in lista:
    if numero in dicionario:
        dicionario[numero] = dicionario[numero] + 1
    else:
        dicionario[numero] = 1
print()
print(dicionario)
```

```python
# Exercicio 3


def lancarDados():
    return randint(1,6) + randint(1,6)

print(lancarDados())

lista = []
for i in range (100000):
    lista.append(lancarDados())
print(lista)

dicionario = {}

for numero in lista:
    if numero in dicionario:
        dicionario[numero] = dicionario[numero] + 1
    else:
        dicionario[numero] = 1
print()
print(dicionario)

dicionario = dict(sorted(dicionario.items()))

print()

for chave, valor in dicionario.items():
    print(f"{chave} : {valor/100000:.2%}")
```
