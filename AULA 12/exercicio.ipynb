```python
# Exemplo

arquivo = open ("teste.txt", "w")

for linha in range (1, 101) :
    arquivo.write (f"Linha {linha}\n" )
arquivo.close ()


arquivo = open("teste.txt", "r")

for linha in arquivo.readlines():
    print(linha)
arquivo.close ()
```

```python
# Exemplo 1

pares = open("pares.txt","w")
impares = open("impares.txt", "w")

for numero in range (1000):
    if numero %2 == 0:
        pares.write(f"{numero}\n")
    else:
       impares.write(f"{numero}\n")
pares.close()
impares.close()
```

```python
# Exemplo 2

with open("pares.txt", "r") as pares, open("multiplos.txt", "w") as multiplos:
    for linha in pares.readlines():
        valor = int(linha.strip())
        if valor % 4 == 0:
            multiplos.write(linha)
    
    
```

```python
# exemplo 3

while True:
    nome = input("Digite o nome: ")
    telefone = input("Digite o telefone: ")
    if nome == "" or telefone == "":
        break

    # Grava o contato dentro do loop
    with open("contatos.txt", "a") as contatos:
        contatos.write(f"{nome}, {telefone}\n")

# Agora lÃª e exibe todos os contatos salvos
with open("contatos.txt", "r") as contatos:
    for linha in contatos:
        nome, telefone = linha.strip().split(",")
        print(f"Nome: {nome:>25} - Telefone: {telefone:<15}")
```

```python
# Exemplo 4

with open("pares.txt", "r") as pares, open("invertido.txt", "w") as invertido:
    linhas = pares.readlines()
    for linha in reversed(linhas):
        invertido.write(linha)
```
